```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.yna.co.kr/view/AKR20220124132600051'   
resp = requests.get(url)
resp
```

```python
html = resp.content
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
tags = soup.select('p')
```

```python
type(tags), len(tags)
```

```python
for i in tags:
    print(i)
```

```python
text = []

for i in tags:
    text.append(i)
```

```python
text
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(text)
```

```python
df
```

```python
df.to_csv('content.csv', encoding='CP949')
```
